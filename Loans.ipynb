```python
## Linear Regression Starts
```

```python
## Creating dummy variables of catagorical data
df_cat=df.select_dtypes(include='object').columns
```

```python
df_cat
```

```python
df_dummies = pd.get_dummies(df[df_cat])
```

```python
df_dummies
```

```python
## numerical columns
df_num=df.select_dtypes(include=np.number).columns
```

```python
df_num
```

```python
df_numerical=df[df_num]
df_numerical
```

```python
## combining the two dataframes

data=pd.concat([df_numerical,df_dummies],axis=1)
```

```python
data
```

```python
data.columns
```

```python
## Splitting the data

X=data.drop(columns=['Interest.Rate','Amount.Requested'])
Y=data['Interest.Rate']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=.7,random_state=100)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ml=LinearRegression()
```

```python
ml.fit(X_train,Y_train)
```

```python
print('Intercept :',ml.intercept_)
```

```python
## Coefficients

print('Coefficient is :',ml.coef_)
```

```python
print(pd.DataFrame(ml.coef_,X_train.columns,columns=['Coefficients']))
```

```python
Y_pred = ml.predict(X_test)
```

```python
Y_pred
```

```python
Y_test-Y_pred
```

```python
df_error=pd.DataFrame(Y_test-Y_pred)
df_error.sort_index(inplace=True)
df_error
```

```python
## R square value
from sklearn import metrics
```

```python
r_square = metrics.r2_score(Y_test,Y_pred)
```

```python
print(r_square)
```

```python
mse=metrics.mean_squared_error(Y_test,Y_pred)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
```

```python
rmse
```

## VIF FACTOR

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# For each X, calculate VIF and save in dataframe
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

## let's try decision tree 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=.7,random_state=100)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=100,max_depth=8)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
df_error=pd.DataFrame(Y_test-Y_pred)
df_error.sort_index(inplace=True)
df_error
```

```python
r_square = metrics.r2_score(Y_test,Y_pred)
```

```python
print(r_square)
```

```python
sns.distplot(Y_pred-Y_test)
```

```python
rmse=np.sqrt(mse)
print(rmse)
```
